```python
element = 'Al'

pr = Project(f'eV_{element}')

bulk = pr.create.structure.bulk(element, cubic=True)

job_relax = pr.create.job.Lammps('relax')
job_relax.structure = bulk
job_relax.calc_minimize(ionic_force_tolerance=0.0001)

job_relax.run()

for strain in np.linspace(-0.1, 0.1, 11):
    
    job = pr.create.job.Lammps(('job', strain))
    relaxed_structure = job_relax.get_structure(-1)
    job.structure = relaxed_structure.apply_strain(strain,return_box=True)
    job.run()
```

```python
table = pr.create.table('eV_table', delete_existing_job=True)
table.add.get_volume
table.add.get_energy_tot
table.add.get_job_name
table.run()
```

```python
df = table.get_dataframe()
df = df.sort_values(by='volume')
df
```
